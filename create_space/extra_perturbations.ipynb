```python
#pops
#p% of audio to max limit
filename='/n/fs/percepaudio/www/mturk_hosts/snr_resize_folder/f1/f1_script1_produced_clipped_10.0.wav'
audio,fs=sf.read(filename)
p=0.1
length=(p/100)*len(audio)
maxi=np.max(audio)
mini=np.min(audio)

A=np.random.choice(int(len(audio)), int(length), replace=False)
#choose randomly amongst these in half
A1=np.random.choice(A,int(len(A)/2),replace=False)
A2=np.setdiff1d(A, A1)
audio[A1]=maxi
audio[A2]=mini
ipd.Audio(audio, rate=fs)
```
